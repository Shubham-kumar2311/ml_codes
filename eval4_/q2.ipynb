```python

def predict(X, w):
    z = np.dot(X, w)
    return (z >= 0).astype(int)

def perceptron_train(X, y, w, epochs=100):
    print(X.shape)
    print(y.shape)
    print(w.shape)
    for _ in range(epochs):
        flag = True
        for i in range(300):
            # z = np.dot(X[i], w)
            # print(w.shape)
            z = predict(X[i],w[i])
            # print(z.shape)
            # print(w.shape)
            y_pred = 1 if z[0] >= 0 else 0
            y_pred = []

            for j in range(300):
                #print(z[j])
                if(z[j] >= 0):
                    y_pred.append(1)
                else:
                    y_pred.append(0)

            y_pred = np.array(y_pred).reshape(300,1)  
            print(y_pred.shape)
            print(y[i])


            # w[i] += (y[i] - y_pred[i]) * X[i]
            if y[i] - y_pred[i] != 0:
                flag = False
        if flag:
            break

       
            # z = np.dot(X[i], w)
            # print(w.shape)
        
        # print(X.shape)
        # print(w.shape)
        z = np.dot(X[i], w)

            # print(z.shape)
            # print(w.shape)
            #y_pred = 1 if z >= 0 else 0
        y_pred = []

        for j in range(300):
                #print(z[j])
            if(z[j] >= 0):
                y_pred.append(1)
            else:
                y_pred.append(0)

        # print(X.shape)
        # print(w.shape)

        y_pred = np.array(y_pred).reshape(300,1) 

        y = y.reshape(300,1)
        print(y_pred.shape)
        # print(X.T.shape)
        # print(X.shape)
        w += np.dot(X.T, (y - y_pred))
        # if y - y_pred != 0:
        #         flag = False
        if flag:
            break

    return w


def perceptron_train(X, y, w_init, epochs):
    w = w_init.copy()
    for _ in range(epochs):
        errors = 0
        for xi, target in zip(X, y):
            y_pred = 1 if np.dot(w, xi) >= 0 else 0
            if y_pred != target:
                errors += 1
                if y_pred == 0 and target == 1:
                    w += xi
                else:
                    w -= xi
        if errors == 0:
            break
    return w
```

```python

def split_data(X, y, train_ratio, val_ratio, test_ratio):
    n = len(y)
    idx = np.random.permutation(n)
    X, y = X[idx], y[idx]
    train_size, val_size = int(train_ratio*n), int(val_ratio*n)
    X_train, y_train = X[:train_size], y[:train_size]
    X_val, y_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
    X_test, y_test = X[train_size+val_size:], y[train_size+val_size:]
    return X_train, y_train, X_val, y_val, X_test, y_test
```

```python
W = np.random.uniform(-0.5, 0.5, (3, 3))
epochs = [50,100,150]
results = []

for i in range(3):
    for ep in epochs:
        w = perceptron_train(X_train, y_train, W[i], ep)
        y_pred = predict(X_test, w)
        results.append({
            "Init_Run": i+1,
            "Epochs": ep,
            "Precision": precision(y_test, y_pred),
            "Recall": recall(y_test, y_pred),
            "Accuracy": accuracy(y_test, y_pred)
        })

df = pd.DataFrame(results)
df.to_csv("result2/perceptron_results.csv", index=False)

for metric in ["Precision","Recall","Accuracy"]:
    plt.figure()
    for run in df["Init_Run"].unique():
        subset = df[df["Init_Run"]==run]
        plt.plot(subset["Epochs"], subset[metric], marker='o', label=f"Init {run}")
    plt.title(metric)
    plt.xlabel("Epochs")
    plt.ylabel(metric)
    plt.legend()
    plt.savefig(f"result2/{metric}.png")
    plt.close()
```
