```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def covariance(df):
    return df['X'].cov(df['Y'])

def correlation(df):
    return df['X'].corr(df['Y'])

def plot(df, title):
    plt.xlabel("X (Independent Variable)")
    plt.ylabel("Y (Dependent Variable)")
    plt.title(title)
    plt.scatter(df['X'], df['Y'])
    plt.show()

# Positive correlation dataset
x_pos = np.random.normal(10, 3, 50)     # avg sd ct
y_pos = 2 * x_pos + np.random.normal(0, 2, 50) 
df_pos = pd.DataFrame({'X': x_pos, 'Y': y_pos})

print("Positive Correlation Example")
print("Covariance:", covariance(df_pos))
print("Correlation:", correlation(df_pos))
plot(df_pos, "Positive Correlation")

# Negative correlation dataset
x_neg = np.random.normal(10, 3, 50)
y_neg = -1.5 * x_neg + np.random.normal(0, 2, 50)
df_neg = pd.DataFrame({'X': x_neg, 'Y': y_neg})

print("\nNegative Correlation Example")
print("Covariance:", covariance(df_neg))
print("Correlation:", correlation(df_neg))
plot(df_neg, "Negative Correlation")
```
