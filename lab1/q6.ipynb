```python
import pandas as pd
import numpy as np

np.random.seed(42)

num_cars = 20
age = np.random.randint(1, 26, num_cars)
mileage = np.random.randint(1, 250, num_cars)

price = 30000 - (age * np.random.randint(800, 1200)) - (mileage * np.random.randint(40, 60)) + np.random.normal(0, 3000, num_cars)

print("Age:", age)
print("Mileage:", mileage)
print("Price:", price)

df = pd.DataFrame({
    'Price': price,
    'Mileage': mileage,
    'Age': age
})

print("Synthetic Car Dataset:\n", df)

#-------------------------#
# Pearson correlation
corr = df.corr()
print("\nPearson Correlation Coefficients:\n", corr)

# Find highest positive and negative correlation
corr_pairs = corr.unstack()
corr_pairs = corr_pairs[corr_pairs != 1]  # ignore self-correlation
highest_pos = corr_pairs.idxmax(), corr_pairs.max()
highest_neg = corr_pairs.idxmin(), corr_pairs.min()

print("\nHighest Positive Correlation:", highest_pos)
print("Highest Negative Correlation:", highest_neg)
```
