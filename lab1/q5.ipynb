```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.random.uniform(1, 10, 20)          
Y = 5 * X + np.random.normal(0, 5, 20)

df = pd.DataFrame({'X': X, 'Y': Y})


plt.scatter(df['X'], df['Y'], color='green')
plt.xlabel("Fertilizers Used (tons)")
plt.ylabel("Crop Production (tons)")
plt.title("Scatter Plot: Fertilizers vs Crop Production")
plt.grid(True)
plt.show()


print("Population Statistics (Entire Dataset):")
population_stats = df.agg(['mean', 'median', 'std', 'var'])
print(population_stats)


mean_crop = df['Y'].mean()
sample_df = df[df['Y'] > mean_crop]

print("\nSample Statistics (Crop Production > Mean):")
sample_stats = sample_df.agg(['mean', 'median', 'std', 'var'])
print(sample_stats)
```
