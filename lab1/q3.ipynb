```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def covariance(df):
    return df['X'].cov(df['Y'])

def correlation(df):
    return df['X'].corr(df['Y'])

def plot(df, title):
    plt.xlabel("X (Independent Variable)")
    plt.ylabel("Y (Dependent Variable)")
    plt.title(title)
    plt.scatter(df['X'], df['Y'])
    plt.show()



# High variance dataset
x_high = np.random.normal(50, 20, 50)   # Mean=50, SD=20 ct=50
y_high = 3 * x_high + np.random.normal(0, 30, 50)
df_high = pd.DataFrame({'X': x_high, 'Y': y_high})

print("High Variance Dataset")
print("Covariance:", covariance(df_high))
print("Correlation:", correlation(df_high))
plot(df_high, "High Variance Dataset")


# Low variance dataset
x_low = np.random.normal(50, 2, 50)    
y_low = 3 * x_low + np.random.normal(0, 2, 50) 
df_low = pd.DataFrame({'X': x_low, 'Y': y_low})

print("\nLow Variance Dataset")
print("Covariance:", covariance(df_low))
print("Correlation:", correlation(df_low))
plot(df_low, "Low Variance Dataset")
```
